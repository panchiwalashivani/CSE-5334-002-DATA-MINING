```python
#import libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

# Load Gender_Classifier.csv dataset

```python
#Load dataset
df = pd.read_csv('Gender_Classifier.csv',index_col=0,encoding='latin-1')
```

# Preprocessing

```python
#check for null values
df.isnull().sum()
```

```python
#drop the columns
df = df.drop(columns = ['description', 'gender_gold', 'name', 'profile_yn_gold', 'profileimage', 'text', 'tweet_coord', 'tweet_location', 'user_timezone', '_last_judgment_at'], axis = 1)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
#filling null values in 'gender:confidence' column
df['gender:confidence'].fillna(df['gender:confidence'].mean(),inplace=True)
```

```python
#for check anyone has null values or not
df.isnull().sum()
```

```python
#Labelencoder
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
```

```python
df['_golden']=le.fit_transform(df['_golden'])
df['_unit_state']=le.fit_transform(df['_unit_state'])
df['profile_yn']=le.fit_transform(df['profile_yn'])
```

```python
le.fit(df['gender'])
le_name_mapping = dict(zip(le.classes_, le.transform(le.classes_)))
print(le_name_mapping)
```

```python
df['gender']=le.fit_transform(df['gender'])
```

```python
df.head()
```

```python
#heatmap of correlation matrix
plt.figure(figsize=(20,8))
sns.heatmap(df.corr(),annot=True)
plt.show()
```

```python
# selecting imporatant comlumns for x and target column to y
x=df[['gender:confidence','profile_yn','profile_yn:confidence','fav_number','retweet_count','tweet_count']]
y=df[['gender']]
```

**Split your dataset 60% for training, 20 for validation and 20% for testing the classifier. Use random state as 2022**

```python
#splitting dataset
X1_train, X_test, y1_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 2022)
X_train, X_val, y_train, y_val = train_test_split(X1_train, y1_train, test_size = 0.2, random_state = 2022)
```

#Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier 
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

Using gini to measure the quality of a split

```python
#tree_1 is the model using "gini" and y_pred_gini is the output class
tree_1 = DecisionTreeClassifier(criterion='gini')
tree_1.fit(X_train, y_train)
y_pred_gini = tree_1.predict(X_test)
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test, y_pred_gini)
print(f"The accuracy of the split using gini is {acc:.2%}")
```

Using entropy to measure the quality of a split

```python
#tree_2 is the model using "entropy" and y_pred_entropy is the output class
tree_2 = DecisionTreeClassifier(criterion='entropy', max_depth=7, min_samples_leaf=4)
tree_2.fit(X_train, y_train)
y_pred_entropy = tree_2.predict(X_test)
from sklearn.metrics import accuracy_score
acc= accuracy_score(y_test, y_pred_entropy) 
print(f"The accuracy of the split using entropy is {acc:.2%}")
```

Calculate and print the confusion matrix for gini and the classification Report.

```python
#confusion mmatrix using graphics 
labelss = tree_1.classes_
cm = confusion_matrix(y_test, y_pred_gini)
ax= plt.subplot()
sns.heatmap(cm, annot=True, fmt='g', ax=ax);
ax.set_xlabel('Pred');ax.set_ylabel('True'); 
ax.set_title('Confusion Matrix'); 
ax.xaxis.set_ticklabels(labelss); ax.yaxis.set_ticklabels(labelss);
```

```python
# print classification report for gini
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred_gini))
```

Calculate and print the confusion matrix for entropy and the classification Report.

```python
#confusion mmatrix using graphics
labelss = tree_2.classes_
cm = confusion_matrix(y_test, y_pred_entropy)
ax= plt.subplot()
sns.heatmap(cm, annot=True, fmt='g', ax=ax);
ax.set_xlabel('Pred');ax.set_ylabel('True'); 
ax.set_title('Confusion Matrix'); 
ax.xaxis.set_ticklabels(labelss); ax.yaxis.set_ticklabels(labelss);
```

```python
# print classification report for entropy
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred_entropy))
```

# Naive Bayes

# Multinomial

```python
# Multinomial Naive bayes classifier
import sklearn.naive_bayes
from sklearn.naive_bayes import MultinomialNB
from sklearn import metrics
classifier = MultinomialNB()
classifier.fit(X_train, y_train.values.ravel())
y_pred_NB = classifier.predict(X_test)
acc = (metrics.accuracy_score(y_test,y_pred_NB))
print(f"The testing accuracy of the model is {acc:.2%}")
```

```python
cf_matrix = confusion_matrix(y_test,y_pred_NB)

print(f'Classification Report:\n{classification_report(y_test,y_pred_NB)}')

df_cm = pd.DataFrame(cf_matrix, columns=np.unique(y_test), index = np.unique(y_test))
df_cm.index.name = 'Actual'
df_cm.columns.name = 'Predicted'

print("Confusion Matrix:\n")
sns.set(font_scale=1.4)
ax=sns.heatmap(df_cm, annot=True,fmt='d')
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
```

#  Visualize the dataset, for the target variable - 2 graphs

```python
plt.figure(figsize=(15,10))

ax = sns.countplot(x = df['profile_yn'], hue = 'gender', palette = 'inferno', data = df)

for bars in ax.containers:
    if bars.get_label() == 0:
        group_total = len(df)
    else:
        group_total = len(df)
    for p in bars.patches:
        percentage = f'{100 * p.get_height() / group_total:.1f}%\n'
        x = p.get_x() + p.get_width() / 2
        y = p.get_height()
        ax.annotate(percentage, (x, y), ha='center', va='center')
plt.show()
```
